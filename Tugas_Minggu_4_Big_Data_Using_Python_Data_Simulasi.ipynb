```python
# Regresi/Prediksi/Peramalan

import time
import pandas as pd
import numpy as np


df = pd.read_csv('climate_data_real.csv')
df
```

```python
#menentukan kolom input x dan output y
x = df.drop(['Year','TemperatureDiff'], axis=1)
y = df.drop(['Year',	'Population',	'CO2_ppm',	'CH4_ppb', 'N20_ppb'], axis=1)
x
```

```python
y
```

Variasi Data Testing 20%

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt

lreg = LinearRegression()

#menentukan data training 70% dan testing 30%
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)

start_time = time.time()

#training
lreg.fit(X_train, y_train)

print('Waktu training:', (time.time() - start_time), 'detik')
```

```python
#testing
start_time = time.time()
y_predicted = lreg.predict(X_test)

mae = metrics.mean_absolute_error(y_test, y_predicted)
mse = metrics.mean_squared_error(y_test, y_predicted)
r2 = metrics.r2_score(y_test, y_predicted)

print('Waktu testing:', (time.time() - start_time), 'detik')

print("Hasil uji akurasi")
print("--------------------------------------")
print('MAE is {}'.format(mae))
print('MSE is {}'.format(mse))
print('R2 score is {}'.format(r2))
```

```python
import pyspark
from pyspark.context import SparkContext
spark = SparkSession\
        .builder\
        .appName("Map Pada Spark")\
        .getOrCreate()
        
#sc = SparkContext.getOrCreate()
sc = spark.sparkContext
sc
```

```python
dataset = spark.read.csv('climate_data_real.csv', header = True, inferSchema = True)
dataset
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
import time
```

```python
dataset = dataset.select(dataset.Population, dataset.CO2_ppm, dataset.CH4_ppb, dataset.N20_ppb,dataset.TemperatureDiff.alias('label'))
dataset.show(10)
```

Variasi Data Testing 20%

```python
#split data training 80% dan testing 20%
training, test = dataset.randomSplit([0.80, 0.20], seed = 100)

# menggabungkan data pada kolom input menjadi satu array bernama features
assembler = VectorAssembler().setInputCols(['Population','CO2_ppm','CH4_ppb','N20_ppb',]).setOutputCol('features')
trainingSet = assembler.transform(training)

# memiih data features dan label untuk training
trainingSet = trainingSet.select("features","label")
trainingSet.show()
```

```python
start = time.time()

# set data ke fungsi linear regression untuk training
lr = LinearRegression()

#training
lr_Model = lr.fit(trainingSet)

testSet = assembler.transform(test)
print('Waktu training:', (time.time() - start), 'detik')
```

```python
from pyspark.ml.evaluation import RegressionEvaluator
evaluator = RegressionEvaluator()

start = time.time()

# memilih kolom features dan label untuk testing
testSet = testSet.select("features", "label")

#testing
testSet = lr_Model.transform(testSet)

# Root Mean Square Error
rmse = evaluator.evaluate(testSet)
print("RMSE: %.3f" % rmse)

# Mean Square Error
mse = evaluator.evaluate(testSet, {evaluator.metricName: "mse"})
print("MSE: %.3f" % mse)

# Mean Absolute Error
mae = evaluator.evaluate(testSet, {evaluator.metricName: "mae"})
print("MAE: %.3f" % mae)

# r2 - coefficient of determination
r2 = evaluator.evaluate(testSet, {evaluator.metricName: "r2"})
print("r2: %.3f" %r2)
print('Waktu testing:', (time.time() - start), 'detik')

testSet.show(10)
```
